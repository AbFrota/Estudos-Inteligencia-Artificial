### Objetivo
#### Prever de acordo com o conjunto de dados se um carro vai ser vendido ou não (sold)
#### Features = millieage_per_year , model_year, price



```python
import pandas as pd

uri = "https://gist.githubusercontent.com/guilhermesilveira/4d1d4a16ccbf6ea4e0a64a38a24ec884/raw/afd05cb0c796d18f3f5a6537053ded308ba94bf7/car-prices.csv"
dados = pd.read_csv(uri)
dados.head()
!pip install graphviz ==0.10
```

```python
## renomeado as colunas para facilitar o entendimento 


a_renomear = { 
    'mileage_per_year' : 'milhas_por_ano',
    'model_year' : 'ano_do_modelo',
    'price' : 'preco',
    'sold' :'vendido'
}
dados = dados.rename(columns=a_renomear)
dados.head()

```

```python
## podemos modificar os valores da coluna 'vendido' para valores '0' e '1'

a_trocar = {
    'no': 0,
    'yes': 1
}
dados.vendido = dados.vendido.map(a_trocar)
dados.head()
```

```python
## Faremos uma alteração, invés de usarmos o ano do modelo é aconselhavel utilizarmos a idade do carro, ou seja
## o calculo desde o ano de fabricação (ano do modelo) até o dia de hoje

from datetime import datetime

ano_atual = datetime.today().year
dados['idade_do_modelo'] = ano_atual - dados.ano_do_modelo
dados.head()
```

```python
## Faremos mais uma alteração, temos a coluna 'milhas_por_ano' que é uma medida válida no Brasil porém como estamos acostumados com KM/ano
## criaremos uma coluna nova que apresentará esses valores em quilômetros (km_por_ano).
## Sabendo que cada milha equivale a 1.60934km
dados ['km_por_ano'] = dados.milhas_por_ano * 1.60934
dados.head()
```

```python
## Logo podemos excluir colunas que não utilizamos mais

dados = dados.drop(columns = ["Unnamed: 0", "milhas_por_ano", "ano_do_modelo"], axis=1)
dados.head()
```

```python
## Hora de separar nossas FEATURES da Classificação

x = dados[["preco", "idade_do_modelo", "km_por_ano"]]
y = dados["vendido"]
```

```python
## Por curiosidade vamos rodar o LinearSVC como algoritmo classificador para modelar como nossos dados serão classificados

import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.svm import LinearSVC
from sklearn.metrics import accuracy_score

x = dados[["preco", "idade_do_modelo", "km_por_ano"]]
y = dados["vendido"]

SEED = 5
np.random.seed(SEED)
train_x, test_x, train_y, test_y = train_test_split(x, y, test_size = 0.25,
                                                         stratify = y)
print("Treinaremos com %d elementos e testaremos com %d elementos" % (len(train_x), len(test_x)))

modelo = LinearSVC()
modelo.fit(train_x, train_y)
previsoes = modelo.predict(test_x)

acuracia = accuracy_score(test_y, previsoes) * 100
print("A acurácia com o LINEARSVC foi %.2f%%" % acuracia)
```

```python
## Criando uma baseline automatizada para utilizar como recurso comparativo 

from sklearn.dummy import DummyClassifier
dummy_stratified = DummyClassifier()
dummy_stratified.fit(train_x, train_y)
acuracia = dummy_stratified.score(test_x, test_y) * 100

print("A acurácia do dummy stratified foi %.2f%%" % acuracia)

## Podemos utilizar Diversas estratégias de classificação, acima utilizamos o padrão do DummyClassifier que é stratified
## Agora vamos ver o que acontece ao utilizar most_frequency

dummy_mostfrequent = DummyClassifier()
dummy_mostfrequent.fit(train_x, train_y)
previsoes = dummy_mostfrequent.predict(test_x)

acuracia = accuracy_score(test_y, previsoes) * 100
print("A acurácia do dummy mostfrequent foi %.2f%%" % acuracia)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.svm import LinearSVC, SVC
from sklearn.metrics import accuracy_score
import numpy as np


SEED = 5
np.random.seed(SEED)
raw_train_x, raw_test_x, train_y, test_y = train_test_split(x, y, test_size = 0.25,
                                                         stratify = y)
print("Treinaremos com %d elementos e testaremos com %d elementos" % (len(train_x), len(test_x)))

## raw_train_x e raw_test_x são as váriaveis com os valores originais, antes da padronização
scaler = StandardScaler()
scaler.fit(raw_train_x)
train_x = scaler.transform(raw_train_x)
test_x = scaler.transform(raw_test_x)

modelo = SVC()
modelo.fit(train_x, train_y)
previsoes = modelo.predict(test_x)

acuracia = accuracy_score(test_y, previsoes) * 100
print("A acurácia foi %.2f%%" % acuracia)
```
